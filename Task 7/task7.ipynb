```python
import tensorflow as tf
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/codebasics/deep-learning-keras-tf-tutorial/refs/heads/master/13_dropout_layer/sonar_dataset.csv')
```

```python
df.sample(5)
```

```python
df['R'] = df['R'].map({'R':1,'M':0})
```

```python
X,y = df.drop(columns='R') , df['R']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape
```

```python
model = keras.Sequential([
    keras.layers.Dense(100,input_shape = (60,) , activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(50 , activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(25 , activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(1 , activation='sigmoid'),
])
model.compile(optimizer = 'adam' , loss = 'binary_crossentropy' , metrics = ['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=500)
```

```python
model.evaluate(X_test,y_test)
```

```python
y_pred = (model.predict(X_test) > 0.5)
```

```python
from sklearn.metrics import classification_report , confusion_matrix
print(classification_report(y_test,y_pred))
```
