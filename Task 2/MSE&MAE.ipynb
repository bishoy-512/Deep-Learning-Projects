```python
import numpy as np
```

```python
y_predicted = np.array([1,1,0,0,1])
y_true = np.array([0.30,0.7,1,0,0.5])
```

## Mean Square Error Implemention

##### 1st way

```python
def mse(y_true,y_pred):
    total_error = 0
    for val1,val2 in zip(y_true,y_pred):
        total_error += (val2-val1)**2
    mse = total_error/len(y_true)
    return mse
```

```python
mse(y_true,y_predicted)
```

##### 2nd Way

```python
def mse(y_true,y_pred):
    mean = np.mean(np.square(y_true-y_predicted))
    return mean
```

```python
mse(y_true,y_predicted)
```
